```python
import json
import pandas as pd
import numpy as np
import ast
import re
from ast import literal_eval
```

```python
df_prueba = pd.read_csv('data.csv')
```

```python
def convert_to_list(x):
        try:
            return literal_eval(x)
        except (ValueError, SyntaxError):
            return []
```

```python
df_prueba['genres']=df_prueba['genres'].apply(convert_to_list)
```

```python
df_prueba['genres'] = df_prueba['genres'].apply(lambda x: x if x != 0 else '')
```

```python
genero_contado={}
for i in df_prueba['genres']:
    for j in i:
        genero_contado[j]= genero_contado.get(j,0)+1
generos_mas_vendidos = sorted(genero_contado , key=genero_contado.get, reverse=True)[:5]           
```
